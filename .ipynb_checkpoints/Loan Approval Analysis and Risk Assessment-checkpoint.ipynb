## 🧹 Data Cleaning

Handling missing values is a crucial step to ensure data quality and model performance.

### 🔍 Missing Values Analysis
We first identify missing values in each column.

### 🛠️ Strategy:
- Categorical → Fill with Mode
- Numerical → Fill with Median/Mean

```python
missing = df.isnull().sum()
missing = missing[missing > 0]

print("Missing Values:\n", missing)
```

```python
# Filling categorical columns with mode
df['Gender'] = df['Gender'].fillna(df['Gender'].mode()[0])
df['Married'] = df['Married'].fillna(df['Married'].mode()[0])
df['Self_Employed'] = df['Self_Employed'].fillna(df['Self_Employed'].mode()[0])
df['Dependents'] = df['Dependents'].fillna(df['Dependents'].mode()[0])
df['Loan_Amount_Term'] = df['Loan_Amount_Term'].fillna(df['Loan_Amount_Term'].mode()[0])
df['Credit_History'] = df['Credit_History'].fillna(df['Credit_History'].mode()[0])
```

```python
# Filling Numerical columns with median

df['LoanAmount'] = df['LoanAmount'].fillna(df['LoanAmount'].median())
```

```python
df.isnull().sum()
```

```python
df['Dependents'] = df['Dependents'].replace('3+', 3).astype(int)
```

```python
df.info()
```

```python
df.duplicated().sum()
```

```python
(df['LoanAmount'] <= 0).sum()
```

## 📊 Exploratory Data Analysis (EDA)

EDA helps us understand patterns, relationships, and trends in the dataset.

```python
plt.figure(figsize=(7,5))

colors = ['#4CAF50', '#F44336']  # green & red

ax = sns.countplot(x='Loan_Status', hue='Loan_Status', data=df, palette=colors, legend=False)

# Percentage labels
total = len(df)
for p in ax.patches:
    percentage = f'{100 * p.get_height()/total:.1f}%'
    ax.annotate(percentage,
                (p.get_x() + p.get_width()/2., p.get_height()),
                ha='center', va='bottom', fontsize=10)

plt.title("Loan Approval Distribution (%)", fontsize=14, fontweight='bold')
plt.xlabel("Loan Status")
plt.ylabel("Count")

plt.show()
```

💡 Insight:
Majority of applicants are approved/rejected — observe the class balance.
This is important for model training.

```python
# 1. Credit History vs Loan Approval

plt.figure(figsize=(7,5))

ax = sns.countplot(x='Credit_History', hue='Loan_Status', data=df, palette='Set2')

plt.title("Credit History vs Loan Status", fontweight='bold')
plt.show()
```

## Insight:
People with credit history = higher approval chances
This is strongest factor

```python
# 2. Total Income vs Loan Approval
df['Total_Income'] = df['ApplicantIncome'] + df['CoapplicantIncome']
```

```python
plt.figure(figsize=(7,5))

sns.boxplot(x='Loan_Status', y='Total_Income', hue='Loan_Status',
            data=df, palette='Set2', legend=False)

sns.stripplot(x='Loan_Status', y='Total_Income',
              data=df, color='black', alpha=0.3)

plt.title("Total Income vs Loan Status", fontweight='bold')
plt.show()
```

## Insights
There is significant overlap in total income between approved and rejected applicants.
This suggests that income alone is not a strong deciding factor for loan approval.
Other features like credit history play a more important role.

```python
# 3. Loan Amount vs Loan Status
plt.figure(figsize=(7,5))

sns.boxplot(x='Loan_Status', y='LoanAmount',
            hue='Loan_Status', data=df,
            palette='coolwarm', legend=False)

plt.title("Loan Amount vs Loan Status", fontweight='bold')
plt.show()
```

# 💡 Insight:
👉 Higher loan → higher risk → lower approval

```python
# 4. Property Area vs Loan Status
sns.countplot(x='Property_Area', hue='Loan_Status', data=df)
plt.title("Property Area vs Loan Status")
plt.show()
```

# 💡 Insight:

```python
# 6. Married vs Loan Approval

sns.countplot(x='Married', hue='Loan_Status', data=df)
plt.title("Married vs Loan Status")
plt.show()
```

```python
# 7. Dependents vs Loan Status
sns.countplot(x='Dependents', hue='Loan_Status', data=df)
plt.title("Dependents vs Loan Status")
plt.show()
```

# 💡 Insight:

```python
# 8. Self Employed vs Loan Status
sns.countplot(x='Self_Employed', hue='Loan_Status', data=df)
plt.title("Self Employed vs Loan Status")
plt.show()
```

```python
# 💡 Insight:
#👉 Salaried vs self-employed risk comparison
sns.boxplot(x='Loan_Status', y='Loan_Amount_Term', data=df)
plt.title("Loan Term vs Loan Status")
plt.show()
```

```python
# 8. Correlation Heatmap
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

# 💡 Insight:
